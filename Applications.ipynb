```python
file_path2 = r'C:\Users\Justa\Desktop\GoIT\Python\industries.csv'
industries = pd.read_csv(file_path2)
```

```python
merged=pd.merge(applications, industries, on='Industry', how='left')
merged[:5]
```

```python
unique_education_levels = merged['Education level'].unique()
print(unique_education_levels)

unique_locations = merged['Location'].unique()
print(unique_locations)
```

```python
gender_translation = {
    'чоловік': 'male',
    'жінка': 'female'
}

education_translation = {
    'вища (бакалавр, спеціаліст, магістр)': 'Higher (Bachelor, Specialist, Master)',
    'ще студент вишу': 'Still a University Student',
    'середня спеціальна': 'Specialized Secondary',
    'середня': 'Secondary',
    'дві вищі': 'Two Higher Educations',
    None: 'Unknown'
}

location_translation = {
    'івано-франківськ чи область': 'Ivano-Frankivsk or Region',
    'київ чи область': 'Kyiv or Region',
    'львів чи область': 'Lviv or Region',
    'житомир чи область': 'Zhytomyr or Region',
    'дніпро чи область': 'Dnipro or Region',
    'тернопіль чи область': 'Ternopil or Region',
    'вінниця чи область': 'Vinnytsia or Region',
    'кропивницький чи область': 'Kropyvnytskyi or Region',
    'чернігів чи область': 'Chernihiv or Region',
    'харків чи область': 'Kharkiv or Region',
    'рівне чи область': 'Rivne or Region',
    'запоріжжя чи область': 'Zaporizhzhia or Region',
    'чернівці чи область': 'Chernivtsi or Region',
    'суми чи область': 'Sumy or Region',
    'одеса чи область': 'Odesa or Region',
    'черкаси чи область': 'Cherkasy or Region',
    'луцьк чи волинська область': 'Lutsk or Volyn Region',
    'ужгород чи закарпатська область': 'Uzhhorod or Zakarpattia Region',
    'полтава чи область': 'Poltava or Region',
    'хмельницький чи область': 'Khmelnytskyi or Region',
    'миколаїв чи область': 'Mykolaiv or Region',
    'херсон чи область': 'Kherson or Region',
    'донецька чи луганська область': 'Donetsk or Luhansk Region',
    None: 'Unknown'  
}

merged['Gender'] = merged['Gender'].str.strip().str.lower()
merged['Education level'] = merged['Education level'].str.strip().str.lower()
merged['Location'] = merged['Location'].str.strip().str.lower()

merged['Gender'] = merged['Gender'].replace(gender_translation)
merged['Education level'] = merged['Education level'].replace(education_translation)
merged['Location'] = merged['Location'].replace(location_translation)
```

```python
merged.info()
```

```python
merged['Applied at'] = pd.to_datetime(merged['Applied at'], format='mixed', dayfirst=True)
```

```python
# Calculate the rating
merged['rating'] = 0

# Criterion 1: Age between 35 and 55
merged['rating'] += ((merged['Age'] >= 35) & (merged['Age'] <= 55)) * 20

# Criterion 2: Application submitted during the week (Monday to Friday)
merged['rating'] += (merged['Applied at'].dt.weekday < 5) * 20

# Criterion 3: Marital status is 'Married'
merged['rating'] += (merged['Marital status'] == 'Married') * 20

# Criterion 4: Location is 'Warsaw' or 'Masovian Voivodeship'
merged['rating'] += merged['Location'].isin(['Warsaw', 'Masovian Voivodeship']) * 10

# Criterion 5: Score from the industries table
merged['rating'] += merged['Score']

# Criterion 6: External Rating >= 7
merged['rating'] += (merged['External Rating'] >= 7) * 20

# Criterion 7: External Rating <= 2 (subtract 20 points)
merged['rating'] -= (merged['External Rating'] <= 2) * 20

# Set rating to 0 if 'Amount' is NaN or 'External Rating' is 0
merged.loc[(merged['Amount'].isna()) | (merged['External Rating'] == 0), 'rating'] = 0

# Ensure rating is between 0 and 100
merged['rating'] = merged['rating'].clip(lower=0, upper=100)
```

```python
merged[12000:13500]
```

```python
merged['Week'] = merged['Applied at'].dt.to_period('W').astype(str)
```

```python
accepted_applications = merged.loc[merged['rating'] > 0].copy()
weekly_avg_rating = accepted_applications.groupby('Week')['rating'].mean().reset_index()
```

```python
weekly_avg_rating.info
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
plt.plot(weekly_avg_rating['Week'], weekly_avg_rating['rating'], marker='.', linestyle='-', color='darkcyan')
plt.title('Average Rating of Accepted Applications by Week')
plt.xlabel('Week')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig("Average_Rating_of_Accepted_Applications_by_Week.png")
plt.show()
```
